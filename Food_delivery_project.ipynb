```python
import pandas as pd

orders = pd.read_csv("orders.csv")
orders.head()
```

```python
# STEP 2: Load Users Data (JSON)

import pandas as pd

# Load users dataset
users = pd.read_json("users.json")

# Basic inspection
print("Users Shape:", users.shape)
print("\nUsers Columns:")
print(users.columns)

print("\nUsers Preview:")
users.head()
```

```python
# STEP 3: Load Restaurant Data (SQL)

import sqlite3
import pandas as pd
import os

# Check if the file exists
file_path = "restaurants.sql"
if not os.path.exists(file_path):
    print(f"Error: The file '{file_path}' does not exist.")
else:
    # Connect to an in-memory SQLite database
    conn = sqlite3.connect(":memory:")

    try:
        # Read SQL commands from the .sql file
        with open(file_path, 'r') as f:
            sql_script = f.read()

        # Execute the SQL script to create tables and insert data
        conn.executescript(sql_script)

        # Load restaurant table from the in-memory database
        restaurants = pd.read_sql_query("SELECT * FROM restaurants", conn)

        # Basic inspection
        print("Restaurants Shape:", restaurants.shape)
        print("\nRestaurants Columns:")
        print(restaurants.columns)

        print("\nRestaurants Preview:")
        print(restaurants.head())

    except Exception as e:
        print(f"An error occurred while processing the SQL file: {e}")
    finally:
        # Close connection
        conn.close()
```

```python
#I validated primary-foreign key relationships before performing joins.‚Äù

# STEP 4: Validate Join Keys

# Check if join keys exist
assert "user_id" in orders.columns
assert "user_id" in users.columns
assert "restaurant_id" in orders.columns
assert "restaurant_id" in restaurants.columns

print("All join keys are present.")
```

```python
# STEP 5: Merge Datasets (Left Join)

# Merge orders with users
orders_users = pd.merge(
    orders,
    users,
    on="user_id",
    how="left"
)

# Merge with restaurants
final_dataset = pd.merge(
    orders_users,
    restaurants,
    on="restaurant_id",
    how="left"
)

print("Final Dataset Shape:", final_dataset.shape)
final_dataset.head()
```

```python
# STEP 6: Export Final Dataset

final_dataset.to_csv("final_food_delivery_dataset.csv", index=False)

print("Final dataset saved as final_food_delivery_dataset.csv")
```
